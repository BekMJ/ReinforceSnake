```python
import pygame
import random

# Initialize pygame
pygame.init()

# Colors and dimensions
WHITE = (255, 255, 255)
GREEN = (0, 128, 0)
RED = (255, 0, 0)
BLACK = (0, 0, 0)

# Game dimensions
WIDTH, HEIGHT = 800, 600
CELL_SIZE = 20

# Directions
LEFT = (-1, 0)
RIGHT = (1, 0)
UP = (0, -1)
DOWN = (0, 1)

# Initialize screen and clock
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Snake Game")
clock = pygame.time.Clock()

# Q-table initialization
state_space = (WIDTH // CELL_SIZE) * (HEIGHT // CELL_SIZE)
action_space = 4  # LEFT, RIGHT, UP, DOWN
Q = {}
for x in range(WIDTH // CELL_SIZE):
    for y in range(HEIGHT // CELL_SIZE):
        Q[(x, y)] = [0] * action_space

# Hyperparameters
alpha = 0.1
gamma = 0.9
epsilon = 0.1

# Mapping actions to indices and vice-versa
actions = [LEFT, RIGHT, UP, DOWN]
action_to_idx = {LEFT: 0, RIGHT: 1, UP: 2, DOWN: 3}

def reset_game():
    global snake, snake_dir, food
    snake = [(5, 5), (5, 6), (5, 7)]
    snake_dir = RIGHT
    food = (random.randint(0, (WIDTH // CELL_SIZE) - 1), random.randint(0, (HEIGHT // CELL_SIZE) - 1))

game_active = True

# Main game loop
while game_active: # Outer loop for game repetition
    snake = [(5, 5), (5, 6), (5, 7)]
    snake_dir = RIGHT
    food = (random.randint(0, (WIDTH // CELL_SIZE) - 1), random.randint(0, (HEIGHT // CELL_SIZE) - 1))
    
    running = True
    while running:
        screen.fill(BLACK)
        
        # RL-based action selection
        if random.uniform(0, 1) < epsilon:
            action = random.choice(actions)
        else:
            action_idx = Q[food].index(max(Q[food]))
            action = actions[action_idx]
        
        # Update direction
        snake_dir = action

        # Movement and game logic
        head = snake[0]
        new_x, new_y = head[0] + snake_dir[0], head[1] + snake_dir[1]
        snake = [(new_x, new_y)] + snake[:-1]

        # Check collision with food and game logic
        reward = 0
        if snake[0] == food:
            reward = 1
            snake.append(snake[-1])
            food = (random.randint(0, (WIDTH // CELL_SIZE) - 1), random.randint(0, (HEIGHT // CELL_SIZE) - 1))

        # Draw snake and food
        for segment in snake:
            pygame.draw.rect(screen, GREEN, (segment[0] * CELL_SIZE, segment[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE))

        pygame.draw.rect(screen, RED, (food[0] * CELL_SIZE, food[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE))

        # Check for game over conditions
        if (snake[0][0] < 0 or snake[0][0] >= WIDTH // CELL_SIZE or
                snake[0][1] < 0 or snake[0][1] >= HEIGHT // CELL_SIZE or
                snake[0] in snake[1:]):
            reward = -1
            running = False

        current_q_value = Q[food][action_to_idx[action]]
        next_max_q_value = max(Q[food])
        Q[food][action_to_idx[action]] = current_q_value + alpha * (reward + gamma * next_max_q_value - current_q_value)

        pygame.display.flip()
        clock.tick(10)

    # Game over screen
    screen.fill(BLACK)
    font = pygame.font.SysFont(None, 55)
    text = font.render('Game Over', True, WHITE)
    screen.blit(text, (WIDTH // 2 - text.get_width() // 2, HEIGHT // 2 - text.get_height() // 2))
    prompt_text = font.render('Play again? (Y/N)', True, WHITE)
    screen.blit(prompt_text, (WIDTH // 2 - prompt_text.get_width() // 2, HEIGHT // 2 + text.get_height()))
    pygame.display.flip()

    awaiting_user_decision = True
    while awaiting_user_decision:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_active = False
                awaiting_user_decision = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_y:
                    reset_game()
                    awaiting_user_decision = False
                elif event.key == pygame.K_n:
                    game_active = False
                    awaiting_user_decision = False

pygame.quit()

```
